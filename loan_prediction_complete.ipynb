```python
# Columns to remove
columns_to_remove = []

# Remove ID columns (not useful for prediction)
id_columns = ['Unnamed: 0', 'id', 'member_id', 'url']
columns_to_remove.extend([col for col in id_columns if col in df.columns])

# Remove columns with >50% missing values (updated from 90%)
high_missing_cols = missing_data[missing_data['Missing_Percentage'] > 50]['Column'].tolist()
columns_to_remove.extend(high_missing_cols)

# Remove post-loan columns (data leakage - available only after loan is issued)
post_loan_columns = [
    'funded_amnt', 'funded_amnt_inv', 'out_prncp', 'out_prncp_inv',
    'total_pymnt', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int',
    'total_rec_late_fee', 'recoveries', 'collection_recovery_fee',
    'last_pymnt_d', 'last_pymnt_amnt', 'next_pymnt_d', 'last_credit_pull_d',
    'collections_12_mths_ex_med', 'policy_code', 'acc_now_delinq',
    'tot_coll_amt', 'tot_cur_bal', 'total_rev_hi_lim',
    'issue_d'  # Issue date is post-loan information
]
columns_to_remove.extend([col for col in post_loan_columns if col in df.columns])

# Remove redundant columns
redundant_columns = ['emp_title', 'title', 'zip_code', 'pymnt_plan']
columns_to_remove.extend([col for col in redundant_columns if col in df.columns])

# Remove original loan_status (we have binary version)
if 'loan_status' in df.columns:
    columns_to_remove.append('loan_status')

# Remove duplicates from list
columns_to_remove = list(set(columns_to_remove))
# Print detailed breakdown
print(f"\nColumns to remove: {len(columns_to_remove)}")
print("\n" + "-"*80)
print("Categories of columns being removed:")
print("-"*80)

id_count = len([col for col in id_columns if col in columns_to_remove])
missing_count = len([col for col in high_missing_cols if col in columns_to_remove])
leakage_count = len([col for col in post_loan_columns if col in columns_to_remove])
redundant_count = len([col for col in redundant_columns if col in columns_to_remove])

print(f"  1. ID columns: {id_count}")
if id_count > 0:
    print(f"     {[col for col in id_columns if col in columns_to_remove]}")

print(f"\n  2. High missing (>50%): {missing_count}")
if missing_count > 0:
    print(f"     (First 5): {[col for col in high_missing_cols if col in columns_to_remove][:5]}")
    if missing_count > 5:
        print(f"     ... and {missing_count - 5} more")

print(f"\n  3. Post-loan (data leakage): {leakage_count}")
if leakage_count > 0:
    print(f"     (First 5): {[col for col in post_loan_columns if col in columns_to_remove][:5]}")
    if leakage_count > 5:
        print(f"     ... and {leakage_count - 5} more")

print(f"\n  4. Redundant: {redundant_count}")
if redundant_count > 0:
    print(f"     {[col for col in redundant_columns if col in columns_to_remove]}")

print(f"\n  5. Original target: {'loan_status' if 'loan_status' in columns_to_remove else 'N/A'}")

print("-"*80)

# Create cleaned dataframe
df_clean = df.drop(columns=columns_to_remove, errors='ignore')

print(f"\n Shape Summary:")
print(f"  Original shape: {df.shape}")
print(f"  Cleaned shape:  {df_clean.shape}")
print(f"  Columns removed: {df.shape[1] - df_clean.shape[1]}")
print(f"  Rows retained: {df_clean.shape[0]:,} (100%)")

# Verify target column exists
if target_col in df_clean.columns:
    print(f"\n Target column '{target_col}' retained")
    print(f"   Distribution: {dict(df_clean[target_col].value_counts())}")
else:
    print(f"\nTarget column '{target_col}' not found!")

print("\n Unnecessary columns removed!")

# Optional: Save list of removed columns for reference
removed_cols_df = pd.DataFrame({
    'Removed_Column': columns_to_remove,
    'Reason': ['ID' if col in id_columns else
               'High Missing' if col in high_missing_cols else
               'Data Leakage' if col in post_loan_columns else
               'Redundant' if col in redundant_columns else
               'Original Target' for col in columns_to_remove]
})
```

```python
# Convert revol_util from string to float (remove % sign)
if 'revol_util' in df_clean.columns:
    if df_clean['revol_util'].dtype == 'object':
        print("\nConverting 'revol_util' from string to numeric...")
        df_clean['revol_util'] = df_clean['revol_util'].str.replace('%', '').replace('', np.nan)
        df_clean['revol_util'] = pd.to_numeric(df_clean['revol_util'], errors='coerce')
        print("✓ revol_util converted to numeric")

# Convert int_rate from string to float (remove % sign)
if 'int_rate' in df_clean.columns:
    if df_clean['int_rate'].dtype == 'object':
        print("\nConverting 'int_rate' from string to numeric...")
        df_clean['int_rate'] = df_clean['int_rate'].str.replace('%', '').replace('', np.nan)
        df_clean['int_rate'] = pd.to_numeric(df_clean['int_rate'], errors='coerce')
        print("✓ int_rate converted to numeric")

# Convert term to numeric (extract number of months)
if 'term' in df_clean.columns:
    if df_clean['term'].dtype == 'object':
        print("\nConverting 'term' from string to numeric...")
        df_clean['term'] = df_clean['term'].str.extract('(\d+)').astype(float)
        print("✓ term converted to numeric (months)")
```

```python
# Strategy for handling missing values
imputation_strategy = {}

# 1. For credit history features - missing often means "no history"
credit_history_cols = [
    'mths_since_last_delinq', 'mths_since_last_record', 
    'mths_since_recent_inq', 'mths_since_recent_revol_delinq',
    'mths_since_recent_bc', 'mths_since_recent_bc_dlq'
]

for col in credit_history_cols:
    if col in df_clean.columns and df_clean[col].isnull().sum() > 0:
        # Fill with a large number (e.g., 999) to indicate "no history"
        df_clean[col].fillna(999, inplace=True)
        imputation_strategy[col] = 'Filled with 999 (no history)'
        print(f"✓ {col}: Filled with 999 (no history)")

# 2. For numerical features - use median
numerical_cols_to_impute = ['annual_inc', 'dti', 'revol_util', 'bc_open_to_buy', 
                             'bc_util', 'mo_sin_old_il_acct', 'mths_since_recent_inq',
                             'num_tl_120dpd_2m', 'percent_bc_gt_75', 'tot_hi_cred_lim']

for col in numerical_cols_to_impute:
    if col in df_clean.columns and df_clean[col].isnull().sum() > 0:
        median_value = df_clean[col].median()
        df_clean[col].fillna(median_value, inplace=True)
        imputation_strategy[col] = f'Filled with median ({median_value:.2f})'
        print(f"{col}: Filled with median ({median_value:.2f})")

# 3. For categorical features - use mode or "Unknown"
categorical_cols_to_impute = ['emp_length', 'emp_title', 'title']

for col in categorical_cols_to_impute:
    if col in df_clean.columns and df_clean[col].isnull().sum() > 0:
        mode_value = df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown'
        df_clean[col].fillna(mode_value, inplace=True)
        imputation_strategy[col] = f'Filled with mode ({mode_value})'
        print(f"{col}: Filled with mode/Unknown")

# 4. For remaining columns with <5% missing - drop rows
remaining_missing = df_clean.columns[df_clean.isnull().any()].tolist()
if remaining_missing:
    missing_pct = (df_clean[remaining_missing].isnull().sum() / len(df_clean)) * 100
    cols_to_drop_rows = missing_pct[missing_pct < 5].index.tolist()
    
    if cols_to_drop_rows:
        print(f"\nDropping rows with missing values in: {cols_to_drop_rows}")
        before_drop = len(df_clean)
        df_clean = df_clean.dropna(subset=cols_to_drop_rows)
        print(f"Rows dropped: {before_drop - len(df_clean):,}")

print(f"\nMissing value imputation complete!")
print(f"Remaining missing values: {df_clean.isnull().sum().sum():,}")
print(f"Current dataframe shape: {df_clean.shape}")
```

```python
# Identify numerical columns
numerical_features = df_clean.select_dtypes(include=[np.number]).columns.tolist()

# Remove target variable from the list
numerical_features = [col for col in numerical_features if col not in 
                      ['loan_status_binary']]

print(f"\nAnalyzing outliers in {len(numerical_features)} numerical features...")

outlier_summary = []

for feature in numerical_features:
    Q1 = df_clean[feature].quantile(0.25)
    Q3 = df_clean[feature].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 3 * IQR  # Using 3*IQR for extreme outliers
    upper_bound = Q3 + 3 * IQR
    
    outliers = df_clean[(df_clean[feature] < lower_bound) | 
                        (df_clean[feature] > upper_bound)][feature]
    
    outlier_count = len(outliers)
    outlier_pct = (outlier_count / len(df_clean)) * 100
    
    if outlier_count > 0:
        outlier_summary.append({
            'Feature': feature,
            'Outlier_Count': outlier_count,
            'Outlier_Percentage': outlier_pct,
            'Lower_Bound': lower_bound,
            'Upper_Bound': upper_bound
        })

outlier_df = pd.DataFrame(outlier_summary).sort_values('Outlier_Percentage', ascending=False)

print(f"\nFeatures with outliers: {len(outlier_df)}")
if len(outlier_df) > 0:
    print("\nTop 10 features with most outliers:")
    print(outlier_df.head(10).to_string(index=False))

# Cap extreme outliers for specific features
features_to_cap = ['annual_inc', 'dti', 'revol_bal']

for feature in features_to_cap:
    if feature in df_clean.columns:
        # Cap at 99th percentile
        upper_cap = df_clean[feature].quantile(0.99)
        outliers_before = (df_clean[feature] > upper_cap).sum()
        
        df_clean[feature] = df_clean[feature].clip(upper=upper_cap)
        
        print(f"\n✓ {feature}: Capped {outliers_before:,} values at 99th percentile ({upper_cap:.2f})")
```

```python
# Credit Utilization Ratio
if 'revol_bal' in df_clean.columns and 'revol_util' in df_clean.columns:
    df_clean['credit_utilization'] = df_clean['revol_util']

# Income to Loan Ratio
if 'annual_inc' in df_clean.columns and 'loan_amnt' in df_clean.columns:
    df_clean['income_to_loan_ratio'] = df_clean['annual_inc'] / (df_clean['loan_amnt'] + 1)

# Installment to Income Ratio
if 'installment' in df_clean.columns and 'annual_inc' in df_clean.columns:
    df_clean['installment_to_income'] = (df_clean['installment'] * 12) / (df_clean['annual_inc'] + 1)

# Average FICO Score
if 'fico_range_low' in df_clean.columns and 'fico_range_high' in df_clean.columns:
    df_clean['fico_score'] = (df_clean['fico_range_low'] + df_clean['fico_range_high']) / 2

# Total Credit Lines
if 'open_acc' in df_clean.columns and 'total_acc' in df_clean.columns:
    df_clean['closed_acc'] = df_clean['total_acc'] - df_clean['open_acc']

# Employment Length (convert to numeric)
if 'emp_length' in df_clean.columns:
    emp_length_map = {
        '< 1 year': 0,
        '1 year': 1,
        '2 years': 2,
        '3 years': 3,
        '4 years': 4,
        '5 years': 5,
        '6 years': 6,
        '7 years': 7,
        '8 years': 8,
        '9 years': 9,
        '10+ years': 10
    }
    df_clean['emp_length_numeric'] = df_clean['emp_length'].map(emp_length_map)
    print("Employment length converted to numeric")

# Loan Amount Categories
if 'loan_amnt' in df_clean.columns:
    df_clean['loan_amount_category'] = pd.cut(df_clean['loan_amnt'], 
                                               bins=[0, 10000, 20000, 40000],
                                               labels=['Low', 'Medium', 'High'])
    print("Loan amount categories created")

print(f"New features created: 7")
print(f"Current shape: {df_clean.shape}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (precision_score, recall_score, f1_score, 
                             accuracy_score, confusion_matrix, roc_curve, 
                             precision_recall_curve, roc_auc_score)

print("\n[1] Generating probability predictions...")

# Get probability predictions for validation and test sets
y_val_proba = lr_model.predict_proba(X_val)[:, 1]  # Probability of bad loan
y_test_proba = lr_model.predict_proba(X_test)[:, 1]

print(f"Validation probabilities generated: {len(y_val_proba):,} samples")
print(f"Test probabilities generated: {len(y_test_proba):,} samples")

print("\n[2] Testing multiple threshold values...")

# Define thresholds to test
thresholds = [0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7]

# Store results
results = []

for threshold in thresholds:
    # Make predictions with current threshold
    y_test_pred_thresh = (y_test_proba >= threshold).astype(int)
    
    # Calculate metrics for test set (critical for deployment evaluation)
    test_accuracy = accuracy_score(y_test, y_test_pred_thresh)
    test_precision = precision_score(y_test, y_test_pred_thresh)
    test_recall = recall_score(y_test, y_test_pred_thresh)
    test_f1 = f1_score(y_test, y_test_pred_thresh)
    
    # Get confusion matrix for test set
    cm_test = confusion_matrix(y_test, y_test_pred_thresh)
    tn, fp, fn, tp = cm_test.ravel()
    
    # Calculate error rates
    fpr = fp / (fp + tn)  # False Positive Rate
    fnr = fn / (fn + tp)  # False Negative Rate
    
    # Store results
    results.append({
        'Threshold': threshold,
        'Test_Accuracy': test_accuracy,
        'Test_Precision': test_precision,
        'Test_Recall': test_recall,
        'Test_F1': test_f1,
        'Test_TN': tn,
        'Test_FP': fp,
        'Test_FN': fn,
        'Test_TP': tp,
        'Test_FPR': fpr,
        'Test_FNR': fnr
    })

# Create DataFrame
df_results = pd.DataFrame(results)

print(f"Tested {len(thresholds)} different thresholds")

print("TEST SET PERFORMANCE ACROSS DIFFERENT THRESHOLDS")

# Display key metrics
display_df = df_results[['Threshold', 'Test_Precision', 'Test_Recall', 
                          'Test_F1', 'Test_Accuracy', 'Test_FP', 'Test_FN']].copy()
display_df.columns = ['Threshold', 'Precision', 'Recall', 'F1-Score', 
                       'Accuracy', 'False Pos.', 'False Neg.']

# Format for better display
pd.options.display.float_format = '{:.4f}'.format
print(display_df.to_string(index=False))

# 1. Maximum F1-Score (balanced approach)
best_f1_idx = df_results['Test_F1'].idxmax()
best_f1_thresh = df_results.loc[best_f1_idx, 'Threshold']

print(f"\n1. BALANCED APPROACH (Maximize F1-Score):")
print(f"   Threshold:        {best_f1_thresh:.2f}")
print(f"   Precision:        {df_results.loc[best_f1_idx, 'Test_Precision']:.2%}")
print(f"   Recall:           {df_results.loc[best_f1_idx, 'Test_Recall']:.2%}")
print(f"   F1-Score:         {df_results.loc[best_f1_idx, 'Test_F1']:.4f}")
print(f"   False Positives:  {df_results.loc[best_f1_idx, 'Test_FP']:.0f}")
print(f"   False Negatives:  {df_results.loc[best_f1_idx, 'Test_FN']:.0f}")

# 2. Maximum Precision (minimize false positives)
best_prec_idx = df_results['Test_Precision'].idxmax()
best_prec_thresh = df_results.loc[best_prec_idx, 'Threshold']

print(f"\n2. PRECISION-FOCUSED (Minimize rejecting good applicants):")
print(f"   Threshold:        {best_prec_thresh:.2f}")
print(f"   Precision:        {df_results.loc[best_prec_idx, 'Test_Precision']:.2%}")
print(f"   Recall:           {df_results.loc[best_prec_idx, 'Test_Recall']:.2%}")
print(f"   False Positives:  {df_results.loc[best_prec_idx, 'Test_FP']:.0f}")
print(f"   False Negatives:  {df_results.loc[best_prec_idx, 'Test_FN']:.0f}")

# 3. Custom: Precision >= 75% with highest recall
precision_75_df = df_results[df_results['Test_Precision'] >= 0.75]
if len(precision_75_df) > 0:
    best_custom_idx = precision_75_df['Test_Recall'].idxmax()
    best_custom_thresh = df_results.loc[best_custom_idx, 'Threshold']
    
    print(f"\n3. CUSTOM TARGET (Precision >= 75%, maximize recall):")
    print(f"   Threshold:        {best_custom_thresh:.2f}")
    print(f"   Precision:        {df_results.loc[best_custom_idx, 'Test_Precision']:.2%}")
    print(f"   Recall:           {df_results.loc[best_custom_idx, 'Test_Recall']:.2%}")
    print(f"   False Positives:  {df_results.loc[best_custom_idx, 'Test_FP']:.0f}")
    print(f"   False Negatives:  {df_results.loc[best_custom_idx, 'Test_FN']:.0f}")
    print(f"   Recommended when minimum precision requirement is 75%")
else:
    print(f"\n3. CUSTOM TARGET (Precision >= 75%):")
    print(f"   Maximum precision available: {df_results['Test_Precision'].max():.2%}")


print("CURRENT (0.5) vs RECOMMENDED COMPARISON")

current_idx = df_results[df_results['Threshold'] == 0.5].index[0]

print(f"\nCURRENT THRESHOLD (0.5):")
print(f"  Precision:        {df_results.loc[current_idx, 'Test_Precision']:.2%}")
print(f"  Recall:           {df_results.loc[current_idx, 'Test_Recall']:.2%}")
print(f"  False Positives:  {df_results.loc[current_idx, 'Test_FP']:.0f}")
print(f"  False Negatives:  {df_results.loc[current_idx, 'Test_FN']:.0f}")

print(f"\nBEST PRECISION THRESHOLD ({best_prec_thresh:.2f}):")
print(f"  Precision:        {df_results.loc[best_prec_idx, 'Test_Precision']:.2%}")
print(f"  Recall:           {df_results.loc[best_prec_idx, 'Test_Recall']:.2%}")
print(f"  False Positives:  {df_results.loc[best_prec_idx, 'Test_FP']:.0f}")
print(f"  False Negatives:  {df_results.loc[best_prec_idx, 'Test_FN']:.0f}")

# Calculate improvements
fp_saved = df_results.loc[current_idx, 'Test_FP'] - df_results.loc[best_prec_idx, 'Test_FP']
prec_gain = (df_results.loc[best_prec_idx, 'Test_Precision'] - 
            df_results.loc[current_idx, 'Test_Precision']) * 100
fn_increase = df_results.loc[best_prec_idx, 'Test_FN'] - df_results.loc[current_idx, 'Test_FN']

print(f"\nIMPROVEMENT SUMMARY:")
print(f"   Saves {fp_saved:.0f} additional good applicants")
print(f"   Precision improves by {prec_gain:.1f} percentage points")
print(f"   Trade-off: Accepts {fn_increase:.0f} additional risky loans")

print("\n[3] Creating visualizations...")

fig = plt.figure(figsize=(20, 12))

# Plot 1: Precision-Recall Trade-off
ax1 = plt.subplot(2, 3, 1)
ax1.plot(df_results['Threshold'], df_results['Test_Precision'], 
         'b-o', linewidth=2, markersize=8, label='Precision')
ax1.plot(df_results['Threshold'], df_results['Test_Recall'], 
         'r-s', linewidth=2, markersize=8, label='Recall')
ax1.axvline(x=0.5, color='gray', linestyle='--', linewidth=2, alpha=0.7, label='Current (0.5)')
ax1.axvline(x=best_prec_thresh, color='green', linestyle='--', linewidth=2, alpha=0.7, 
            label=f'Best Precision ({best_prec_thresh:.2f})')
ax1.axhline(y=0.75, color='orange', linestyle=':', linewidth=1.5, alpha=0.5, label='Target (75%)')
ax1.set_xlabel('Threshold', fontsize=12, fontweight='bold')
ax1.set_ylabel('Score', fontsize=12, fontweight='bold')
ax1.set_title('Precision-Recall Trade-off', fontsize=14, fontweight='bold')
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0.5, 1.0])

# Plot 2: F1-Score
ax2 = plt.subplot(2, 3, 2)
ax2.plot(df_results['Threshold'], df_results['Test_F1'], 
         'g-o', linewidth=2, markersize=8)
ax2.axvline(x=best_f1_thresh, color='red', linestyle='--', linewidth=2, alpha=0.7, 
            label=f'Best F1 ({best_f1_thresh:.2f})')
ax2.axvline(x=0.5, color='gray', linestyle='--', linewidth=2, alpha=0.7, label='Current (0.5)')
ax2.set_xlabel('Threshold', fontsize=12, fontweight='bold')
ax2.set_ylabel('F1-Score', fontsize=12, fontweight='bold')
ax2.set_title('F1-Score Across Thresholds', fontsize=14, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

# Plot 3: False Positives vs False Negatives
ax3 = plt.subplot(2, 3, 3)
ax3.plot(df_results['Threshold'], df_results['Test_FP'], 
         'orange', linewidth=2, marker='o', markersize=8, label='False Positives')
ax3.plot(df_results['Threshold'], df_results['Test_FN'], 
         'red', linewidth=2, marker='s', markersize=8, label='False Negatives')
ax3.axvline(x=0.5, color='gray', linestyle='--', linewidth=2, alpha=0.7, label='Current (0.5)')
ax3.axvline(x=best_prec_thresh, color='green', linestyle='--', linewidth=2, alpha=0.7, 
            label=f'Best Prec ({best_prec_thresh:.2f})')
ax3.set_xlabel('Threshold', fontsize=12, fontweight='bold')
ax3.set_ylabel('Count', fontsize=12, fontweight='bold')
ax3.set_title('False Positives vs False Negatives', fontsize=14, fontweight='bold')
ax3.legend(fontsize=9)
ax3.grid(True, alpha=0.3)

# Plot 4: Accuracy
ax4 = plt.subplot(2, 3, 4)
ax4.plot(df_results['Threshold'], df_results['Test_Accuracy'], 
         'purple', linewidth=2, marker='o', markersize=8)
ax4.axvline(x=0.5, color='gray', linestyle='--', linewidth=2, alpha=0.7, label='Current (0.5)')
ax4.set_xlabel('Threshold', fontsize=12, fontweight='bold')
ax4.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax4.set_title('Accuracy Across Thresholds', fontsize=14, fontweight='bold')
ax4.legend(fontsize=10)
ax4.grid(True, alpha=0.3)
ax4.set_ylim([0.85, 0.95])

# Plot 5: Precision-Recall Curve
ax5 = plt.subplot(2, 3, 5)
precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_test_proba)
ax5.plot(recall_curve, precision_curve, 'b-', linewidth=2)
ax5.axhline(y=0.75, color='orange', linestyle=':', linewidth=1.5, alpha=0.5, label='Target (75%)')
ax5.set_xlabel('Recall', fontsize=12, fontweight='bold')
ax5.set_ylabel('Precision', fontsize=12, fontweight='bold')
ax5.set_title('Precision-Recall Curve', fontsize=14, fontweight='bold')
ax5.legend(fontsize=10)
ax5.grid(True, alpha=0.3)

# Plot 6: ROC Curve
ax6 = plt.subplot(2, 3, 6)
fpr_roc, tpr_roc, _ = roc_curve(y_test, y_test_proba)
roc_auc = roc_auc_score(y_test, y_test_proba)
ax6.plot(fpr_roc, tpr_roc, 'b-', linewidth=2, label=f'ROC (AUC = {roc_auc:.4f})')
ax6.plot([0, 1], [0, 1], 'k--', linewidth=1, alpha=0.5, label='Random')
ax6.set_xlabel('False Positive Rate', fontsize=12, fontweight='bold')
ax6.set_ylabel('True Positive Rate', fontsize=12, fontweight='bold')
ax6.set_title('ROC Curve', fontsize=14, fontweight='bold')
ax6.legend(fontsize=10)
ax6.grid(True, alpha=0.3)

plt.suptitle('Threshold Optimization Analysis - Test Set Performance', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

print("Visualizations complete")

print("\n[4] Creating confusion matrix comparison...")

top_thresholds = [
    (0.5, "Current"),
    (best_f1_thresh, "Best F1"),
    (best_prec_thresh, "Best Precision")
]

fig2, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, (threshold, name) in enumerate(top_thresholds):
    y_pred = (y_test_proba >= threshold).astype(int)
    cm = confusion_matrix(y_test, y_pred)
    
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                ax=axes[idx], cbar=True,
                xticklabels=['Good', 'Bad'],
                yticklabels=['Good', 'Bad'],
                annot_kws={'size': 14, 'weight': 'bold'})
    
    title = f'{name} (threshold={threshold:.2f})\n'
    title += f'Precision: {precision:.2%} | Recall: {recall:.2%} | F1: {f1:.3f}'
    axes[idx].set_title(title, fontsize=11, fontweight='bold')
    axes[idx].set_ylabel('Actual', fontsize=11, fontweight='bold')
    axes[idx].set_xlabel('Predicted', fontsize=11, fontweight='bold')

plt.suptitle('Confusion Matrices Comparison', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("Comparison complete")

print(f"\nRECOMMENDED THRESHOLD: {best_prec_thresh:.2f}")
print(f"   Achieves {df_results.loc[best_prec_idx, 'Test_Precision']:.1%} precision")
print(f"   Saves {fp_saved:.0f} good applicants compared to current threshold")

# Save results for reference
threshold_results = df_results.copy()
```

```python
# ============================================================
# PREDICT 5 RANDOM CUSTOMERS WITH DETAILED ANALYSIS
# ============================================================

import numpy as np
import pandas as pd
import pickle

print("=" * 80)
print("LOAN PREDICTION FOR 5 RANDOM CUSTOMERS")
print("=" * 80)

# Load the saved model
print("\n[1] Loading model...")
with open('loan_prediction_model_corrected.pkl', 'rb') as f:
    model_package = pickle.load(f)

print("Model loaded successfully")
print(f"Model type: {model_package['model_type']}")
print(f"Threshold: {model_package['threshold']}")
print(f"Trained on: {model_package['training_date']}")

# Select 5 random customers from test set
print("\n[2] Selecting 5 random customers from test set...")
random_indices = np.random.choice(len(X_test), 5, replace=False)
customers = X_test.iloc[random_indices].copy()
actual_outcomes = y_test.iloc[random_indices].copy()

print(f"Selected customers at indices: {random_indices.tolist()}")

# Make predictions
print("\n[3] Making predictions...")
customers_scaled = model_package['scaler'].transform(customers)
predictions_proba = model_package['model'].predict_proba(customers_scaled)
predictions_class = (predictions_proba[:, 1] >= model_package['threshold']).astype(int)

# Display results for each customer
print("\n" + "=" * 80)
print("PREDICTION RESULTS FOR EACH CUSTOMER")
print("=" * 80)

for i, idx in enumerate(random_indices):
    print(f"\n{'=' * 80}")
    print(f"CUSTOMER {i+1} (Test Index: {idx})")
    print(f"{'=' * 80}")
    
    # Get customer data
    customer = customers.iloc[i]
    actual = actual_outcomes.iloc[i]
    prob_bad = predictions_proba[i, 1]
    prob_good = predictions_proba[i, 0]
    predicted = predictions_class[i]
    
    # Display key features (top 15 most important)
    print("\nKEY CUSTOMER CHARACTERISTICS:")
    print("-" * 80)
    
    # Get feature importance from model
    feature_importance = pd.DataFrame({
        'feature': model_package['feature_names'],
        'value': customer.values,
        'coefficient': model_package['model'].coef_[0]
    })
    feature_importance['abs_coef'] = np.abs(feature_importance['coefficient'])
    feature_importance = feature_importance.sort_values('abs_coef', ascending=False)
    
    # Show top 15 features
    for _, row in feature_importance.head(15).iterrows():
        feature_name = row['feature']
        value = row['value']
        
        # Format feature name for readability
        if 'fico' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (FICO Score)")
        elif 'debt_settlement' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Debt Settlement)")
        elif 'hardship' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Hardship Program)")
        elif 'grade' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Loan Grade)")
        elif 'installment' in feature_name.lower() and 'income' not in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Payment Amount)")
        elif 'int_rate' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Interest Rate)")
        elif 'annual_inc' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Annual Income)")
        elif 'dti' == feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Debt-to-Income)")
        elif 'loan_amnt' in feature_name.lower():
            print(f"  {feature_name:.<45} {value:>10.2f} (Loan Amount)")
        else:
            print(f"  {feature_name:.<45} {value:>10.2f}")
    
    # Prediction results
    print("\n" + "-" * 80)
    print("PREDICTION RESULTS:")
    print("-" * 80)
    print(f"  Probability of GOOD loan:  {prob_good:>6.2%} ({prob_good*100:.2f}%)")
    print(f"  Probability of BAD loan:   {prob_bad:>6.2%} ({prob_bad*100:.2f}%)")
    print(f"  Decision Threshold:        {model_package['threshold']:.2f}")
    print(f"  Risk Level:                ", end="")
    
    # Risk categorization
    if prob_bad < 0.3:
        print("LOW RISK")
    elif prob_bad < 0.6:
        print("MEDIUM RISK")
    else:
        print("HIGH RISK")
    
    print(f"\n  FINAL DECISION:            ", end="")
    if predicted == 0:
        print("✓ APPROVE LOAN (Good Loan Predicted)")
    else:
        print("✗ REJECT LOAN (Bad Loan Predicted)")
    
    # Actual outcome
    print(f"\n  Actual Outcome:            ", end="")
    if actual == 0:
        print("Customer REPAID the loan")
    else:
        print("Customer DEFAULTED on the loan")
    
    # Correctness
    is_correct = (predicted == actual)
    print(f"  Prediction Accuracy:       ", end="")
    if is_correct:
        print("✓ CORRECT")
        if predicted == 1 and actual == 1:
            print("  (True Positive - Correctly identified bad loan)")
        else:
            print("  (True Negative - Correctly identified good loan)")
    else:
        print("✗ INCORRECT")
        if predicted == 1 and actual == 0:
            print("  (False Positive - Rejected a good customer)")
        else:
            print("  (False Negative - Approved a risky customer)")

# Summary statistics
print("\n" + "=" * 80)
print("SUMMARY OF 5 PREDICTIONS")
print("=" * 80)

correct_predictions = sum(predictions_class == actual_outcomes)
approved = sum(predictions_class == 0)
rejected = sum(predictions_class == 1)
actually_good = sum(actual_outcomes == 0)
actually_bad = sum(actual_outcomes == 1)

print(f"\nDecisions Made:")
print(f"  Approved: {approved}/5 ({approved/5*100:.0f}%)")
print(f"  Rejected: {rejected}/5 ({rejected/5*100:.0f}%)")

print(f"\nActual Outcomes:")
print(f"  Good loans: {actually_good}/5 ({actually_good/5*100:.0f}%)")
print(f"  Bad loans:  {actually_bad}/5 ({actually_bad/5*100:.0f}%)")

print(f"\nAccuracy:")
print(f"  Correct predictions: {correct_predictions}/5 ({correct_predictions/5*100:.0f}%)")

# Average risk score
avg_risk = np.mean(predictions_proba[:, 1])
print(f"\nAverage Risk Score: {avg_risk:.2%}")

# Risk distribution
print("\nRisk Distribution:")
for i in range(5):
    risk_score = predictions_proba[i, 1]
    print(f"  Customer {i+1}: ", end="")
    print("█" * int(risk_score * 50), end="")
    print(f" {risk_score:.1%}")

print("\n" + "=" * 80)
print("PREDICTION COMPLETE")
print("=" * 80)

# Optional: Create a summary DataFrame
summary_df = pd.DataFrame({
    'Customer': [f'Customer {i+1}' for i in range(5)],
    'Test_Index': random_indices,
    'Probability_Bad': predictions_proba[:, 1],
    'Predicted': ['APPROVE' if p == 0 else 'REJECT' for p in predictions_class],
    'Actual': ['REPAID' if a == 0 else 'DEFAULTED' for a in actual_outcomes],
    'Correct': ['✓' if predictions_class[i] == actual_outcomes.iloc[i] else '✗' 
                for i in range(5)]
})

print("\nSUMMARY TABLE:")
print("-" * 80)
print(summary_df.to_string(index=False))
print("-" * 80)
```

```python
# ============================================================
# ANALYZE FICO SCORE BIAS
# ============================================================

import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("=" * 80)
print("ANALYZING FICO SCORE BIAS IN MODEL")
print("=" * 80)

# Load model
with open('loan_prediction_model_corrected.pkl', 'rb') as f:
    model_package = pickle.load(f)

model = model_package['model']
feature_names = model_package['feature_names']
coefficients = model.coef_[0]

# Create feature importance dataframe
feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients,
    'Abs_Coefficient': np.abs(coefficients)
}).sort_values('Abs_Coefficient', ascending=False)

print("\n[1] TOP 20 MOST IMPORTANT FEATURES:")
print("-" * 80)
for i, row in feature_importance.head(20).iterrows():
    direction = "↑ increases" if row['Coefficient'] > 0 else "↓ decreases"
    print(f"{row['Feature']:.<40} {row['Coefficient']:>8.4f} ({direction} risk)")

# Find FICO-related features
print("\n[2] FICO SCORE FEATURES:")
print("-" * 80)
fico_features = feature_importance[feature_importance['Feature'].str.contains('fico', case=False)]
print(fico_features[['Feature', 'Coefficient', 'Abs_Coefficient']].to_string(index=False))

# Calculate FICO importance relative to total
total_importance = feature_importance['Abs_Coefficient'].sum()
fico_importance = fico_features['Abs_Coefficient'].sum()
fico_percentage = (fico_importance / total_importance) * 100

print(f"\nFICO Score Importance: {fico_percentage:.2f}% of total model weight")

# Test predictions with different FICO scores
print("\n[3] TESTING PREDICTIONS WITH VARYING FICO SCORES:")
print("-" * 80)

# Create neutral customer (all features = 0)
neutral_customer = {feature: 0.0 for feature in feature_names}

fico_tests = [-3.0, -2.0, -1.0, 0.0, 1.0, 2.0, 3.0]
results = []

for fico_value in fico_tests:
    # Set FICO score
    test_customer = neutral_customer.copy()
    test_customer['last_fico_range_high'] = fico_value
    test_customer['last_fico_range_low'] = fico_value
    
    # Make prediction
    customer_df = pd.DataFrame([test_customer])
    customer_scaled = model_package['scaler'].transform(customer_df)
    prob_bad = model.predict_proba(customer_scaled)[0][1]
    decision = "REJECTED" if prob_bad >= model_package['threshold'] else "APPROVED"
    
    results.append({
        'FICO_Value': fico_value,
        'Probability_Bad': prob_bad,
        'Decision': decision
    })
    
    print(f"FICO: {fico_value:>5.1f}  →  Bad Loan Prob: {prob_bad:>6.1%}  →  {decision}")

# Now test with GOOD features but BAD FICO
print("\n[4] TEST: GOOD APPLICANT WITH BAD FICO:")
print("-" * 80)

good_applicant = neutral_customer.copy()
good_applicant['annual_inc'] = 2.0          # High income
good_applicant['dti'] = -1.0                # Low debt-to-income
good_applicant['delinq_2yrs'] = -0.4        # No delinquencies
good_applicant['pub_rec'] = -0.4            # No public records
good_applicant['last_fico_range_high'] = -2.0  # BAD FICO
good_applicant['last_fico_range_low'] = -2.0

customer_df = pd.DataFrame([good_applicant])
customer_scaled = model_package['scaler'].transform(customer_df)
prob_bad = model.predict_proba(customer_scaled)[0][1]
decision = "REJECTED" if prob_bad >= model_package['threshold'] else "APPROVED"

print("Good applicant (high income, low debt, no delinquencies) BUT bad FICO:")
print(f"  Result: {decision}")
print(f"  Bad Loan Probability: {prob_bad:.1%}")

# Test with BAD features but GOOD FICO
print("\n[5] TEST: BAD APPLICANT WITH GOOD FICO:")
print("-" * 80)

bad_applicant = neutral_customer.copy()
bad_applicant['annual_inc'] = -2.0          # Low income
bad_applicant['dti'] = 2.0                  # High debt-to-income
bad_applicant['delinq_2yrs'] = 2.0          # Many delinquencies
bad_applicant['pub_rec'] = 2.0              # Public records
bad_applicant['last_fico_range_high'] = 2.0   # GOOD FICO
bad_applicant['last_fico_range_low'] = 2.0

customer_df = pd.DataFrame([bad_applicant])
customer_scaled = model_package['scaler'].transform(customer_df)
prob_bad = model.predict_proba(customer_scaled)[0][1]
decision = "REJECTED" if prob_bad >= model_package['threshold'] else "APPROVED"

print("Bad applicant (low income, high debt, delinquencies) BUT good FICO:")
print(f"  Result: {decision}")
print(f"  Bad Loan Probability: {prob_bad:.1%}")

# Visualization
print("\n[6] Creating visualization...")
fico_values = [r['FICO_Value'] for r in results]
probabilities = [r['Probability_Bad'] for r in results]

plt.figure(figsize=(10, 6))
plt.plot(fico_values, probabilities, 'b-o', linewidth=2, markersize=8)
plt.axhline(y=model_package['threshold'], color='r', linestyle='--', 
            linewidth=2, label=f"Threshold ({model_package['threshold']:.2f})")
plt.axhline(y=0.5, color='gray', linestyle=':', alpha=0.5)
plt.xlabel('FICO Score (Scaled)', fontsize=12, fontweight='bold')
plt.ylabel('Probability of Bad Loan', fontsize=12, fontweight='bold')
plt.title('Model Dependency on FICO Score\n(All other features = 0)', 
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()

# Add decision labels
for fico, prob in zip(fico_values, probabilities):
    decision_label = "REJECT" if prob >= model_package['threshold'] else "APPROVE"
    color = 'red' if decision_label == "REJECT" else 'green'
    plt.text(fico, prob + 0.03, decision_label, ha='center', 
            fontsize=8, color=color, fontweight='bold')

plt.tight_layout()
plt.savefig('fico_bias_analysis.png', dpi=150)
print("✓ Saved: fico_bias_analysis.png")

print("\n" + "=" * 80)
print("CONCLUSION:")
print("=" * 80)
if fico_percentage > 30:
    print(f"⚠️  MODEL IS HEAVILY BIASED TOWARD FICO SCORE ({fico_percentage:.1f}%)")
    print("   FICO score dominates the decision-making process")
elif fico_percentage > 15:
    print(f"⚠️  MODEL HAS MODERATE FICO BIAS ({fico_percentage:.1f}%)")
    print("   FICO score has significant influence but not overwhelming")
else:
    print(f"✓ MODEL HAS BALANCED FEATURE IMPORTANCE ({fico_percentage:.1f}%)")
    print("   FICO score is important but not dominant")

print("=" * 80)
```
